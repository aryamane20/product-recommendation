```python
import pandas as pd

# Load the required data
ratings = pd.read_csv("/Users/aryamane/Downloads/final_ratings_cleaned.csv")
clusters = pd.read_csv("/Users/aryamane/Downloads/labeled_clutering_users.csv")  
products = pd.read_csv("/Users/aryamane/Downloads/products.csv")

# Merge cluster info with ratings
ratings_clustered = ratings.merge(clusters[['user_id', 'cluster']], on='user_id', how='left')

# Aggregate ratings per product per cluster
cluster_product_stats = ratings_clustered.groupby(['cluster', 'product_id']).agg(
    avg_rating=('rating', 'mean'),
    rating_count=('rating', 'count')
).reset_index()

cluster_product_stats = cluster_product_stats.merge(products[['product_id', 'product_name']], on='product_id', how='left')

# Getting top 5 products per cluster
top_n = 5
top_products_by_cluster = (
    cluster_product_stats
    .sort_values(['cluster', 'avg_rating', 'rating_count'], ascending=[True, False, False])
    .groupby('cluster')
    .head(top_n)
)
top_products_by_cluster['avg_rating'] = top_products_by_cluster['avg_rating'].round(1)

top_products_by_cluster.to_csv("/Users/aryamane/Downloads/top_products_with_names.csv", index=False)
top_products_by_cluster.head(15)
```
